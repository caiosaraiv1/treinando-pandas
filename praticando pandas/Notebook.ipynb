- Criando Tabulações com GroupBy

```python
df = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/python_dados/refs/heads/main/Dados/contagem_bicicletas.csv')
```

```python
df.head()
```

```python
# Entre parenteses: a coluna que a gente vai agrupar os dados
# Entre colchetes: a coluna que contem os dados que a gente vai aplicar a estatistica
estatisticas_clima = df.groupby('clima')['contagem'].describe()
estatisticas_clima
```

- Funções que podem ser usadas com o groupby
  
- `count()` para realizar contagem
- `sum()` para soma
- `mean()` para a média
- `median()` para a mediana
- `min()` para o valor mínimo
- `max()` para o valor máximo
- `std()` para o desvio-padrão
- `var()` para a variância

```python
soma_clima = df.groupby('clima')['contagem'].sum()
soma_clima
```

```python
soma_clima = df.groupby('clima')['contagem'].sum().sort_values(ascending=False)
soma_clima
```

```python
soma_clima = df.groupby('clima')['contagem'].agg(['sum', 'mean', 'min', 'max'])
soma_clima
```

```python
df_estacao_clima = df.groupby(['estacao', 'clima'])['contagem'].mean()
df_estacao_clima
```

```python
df_estacao_clima_unstack = df_estacao_clima.unstack()
df_estacao_clima_unstack
```

```python
agregacoes = {
    'contagem': 'sum',
    'temperatura': 'mean',
    'sensacao_termica': 'mean',
    'umidade': 'mean'
}

df_agregacoes = df.groupby(['estacao', 'clima']).agg(agregacoes)
df_agregacoes
```

- Criando Tabuções com Pivot Table

```python
df = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/python_dados/refs/heads/main/Dados/contagem_bicicletas.csv')
```

```python
df.head()
```

```python
pivot_clima = pd.pivot_table(df, values = 'contagem', index='clima').sort_values(by='contagem', ascending=False)
pivot_clima
```

```python
pivot_clima = pd.pivot_table(df, values = 'contagem', index='clima', aggfunc='sum').sort_values(by='contagem', ascending=False)
pivot_clima
```

```python
pivot_clima_estacao = pd.pivot_table(df, values = 'contagem', index=['estacao', 'clima'], aggfunc='mean').sort_values(by='contagem', ascending=False)
pivot_clima_estacao
```

```python
pivot_clima_estacao = pd.pivot_table(df, values = 'contagem', index='estacao', columns ='clima', aggfunc='mean')
pivot_clima_estacao
```

```python
pivot_agg = df.pivot_table(
    index=['estacao', 'clima'],
    values=['contagem', 'temperatura', 'sensacao_termica', 'umidade'],
    aggfunc={
        'contagem': 'sum',
        'temperatura': 'mean',
        'sensacao_termica': 'mean',
        'umidade': 'mean'
    }
)

pivot_agg
```

```python
pivot_table = df.pivot_table(
    index='estacao',
    columns='clima',
    values='contagem',
    aggfunc={'contagem': 'sum'},
    margins=True,
    margins_name='Total',
    fill_value=0
)

pivot_table
```

```python
dados = {
    'Texto': [
        "Bom dia! ☀️ Começando mais uma semana produtiva!",
        "Novo produto chegando! 🎉 #novidade #lancamento",
        "Promoção relâmpago ⚡ Aproveitem! #promo #desconto",
        "Boa noite pessoal! 🌙 Até amanhã! #boanoite",
        "Dica do dia: 💡 Mantenha-se hidratado! #saude #bemestar"
    ],
    'Horário': ['07:30', '15:45', '12:20', '22:15', '10:00'],
    'Hashtags': ['', 'novidade,lancamento', 'promo,desconto', 'boanoite', 'saude,bemestar'],
    'Caracteres_Emoji': ['☀️', '🎉', '⚡', '🌙', '💡']
}

df_posts = pd.DataFrame(dados)
```

```python
def get_periodo(horario):
    hora = int(horario.split(':')[0])
    if 0 <= hora < 6:
        return 'Madrugada'
    elif 6 <= hora < 12:
        return 'Manhã'
    elif 12 <= hora < 18:
        return 'Tarde'
    else:
        return 'Noite'
```

```python
def analisar_conteudo(row):
    num_hashtags = len([tag for tag in row['Hashtags'].split(',') if tag])
    num_emojis = len(row['Caracteres_Emoji'])
    comprimento_texto = len(row['Texto'])
    
    return {
        'num_hashtags': num_hashtags,
        'num_emojis': num_emojis,
        'comprimento_texto': comprimento_texto
    }
```

```python
df_posts['Período_Dia'] = df_posts['Horário'].apply(get_periodo)
df_posts['Análise_Conteúdo'] = df_posts.apply(analisar_conteudo, axis=1)
```
